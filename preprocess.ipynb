```python
import pandas as pd
import os
import re
```

```python
all_files = []
for files in os.listdir("data"):
    if files.endswith(".tsv"):
        filepath = "data" + "/" + files
        all_files.append(filepath)
all_files
```

# Add "Date" column to our dataframe
I am using ReGex to extract the date format in filename

```python
regex_date = "\d{4}-\d{2}-\d{2}"
for file in all_files:
    df = pd.read_csv(file, sep="\t")
    date = re.search(regex_date, file).group()
    df["Date"] = date
    df.to_csv(file[0:5] + "data_processed" + file[4:-4] + "-processed" + file[-4:], sep="\t", index=False)
```

# Merge all files into one larger data file

```python
os.listdir('data/data_processed/')[-14:-1]
```

```python
# f"data/worldometers-2022-05-{d}.tsv"
path = 'data/data_processed/'
frames = []
all_length = []
# we merged data from the last 13 days, from 14-05 to 26-05
for file in os.listdir('data/data_processed/')[-14:-1]:
    if file.endswith('.tsv'):
        filepath = path + file
        df1 = pd.read_csv(filepath, sep="\t")
        frames.append(df1)
        result = pd.concat(frames)
        length_1month = len(df1.index)
        all_length.append(length_1month)
```

```python
result.to_csv('covid_1405_2605.tsv', sep="\t", index=False)
```
