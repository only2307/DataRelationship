```python
import pandas as pd
import os
import re
```

```python
all_files = []
for files in os.listdir("data"):
    if files.endswith(".tsv"):
        filepath = "data" + "/" + files
        all_files.append(filepath)
all_files
```

# Add "Date" column to our dataframe
I am using ReGex to extract the date format in filename

```python
regex_date = "\d{4}-\d{2}-\d{2}"
for file in all_files:
    df = pd.read_csv(file, sep="\t")
    date = re.search(regex_date, file).group()
    df["Date"] = date
    df.to_csv(file[0:5] + "data_processed" + file[4:-4] + "-processed" + file[-4:], sep="\t", index=False)
```

# Merge all files into one larger data file

```python
os.listdir('data/data_processed/')
```

```python
path = 'data/data_processed/'
frames = []
all_length = []
for file in os.listdir(path):
    if file.endswith('.tsv'):
        filepath = path + file
        df1 = pd.read_csv(filepath, sep="\t")
        frames.append(df1)
        result = pd.concat(frames)
        length_1month = len(df1.index)
        all_length.append(length_1month)
```

```python
# to check if the merged file has the same length with all data files we've crawled
assert sum(all_length) == len(pd.read_csv("covid_data.tsv", sep="\t"))
```

```python
result = result.drop(columns=["NewCases", "NewDeaths", "NewRecovered", "Tests/1M pop ", "NewRecovered\t"], axis=1)
result.to_csv('covid_data.tsv', sep="\t", index=False)
```
